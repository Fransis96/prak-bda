```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Load data
data_sales = pd.read_csv("100 Sales Records.csv")
```

2. Pilih fitur numerik untuk analisis

```python
# Ambil kolom numerik
numerical_features = ['Order_ID', 'Units_Sold', 'Unit_Price', 'Unit_Cost',
                      'Total_Revenue', 'Total_Cost', 'Total_Profit']
x = data_sales[numerical_features].values
```

```python
# Normalisasi
x = StandardScaler().fit_transform(x)
```

```python
# PCA jadi 2 dimensi
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

```python
# Visualisasi
plt.figure(figsize=(8,8))
plt.scatter(principal_df['PC1'], principal_df['PC2'], s=50, alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of 100 Sales Records Dataset')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data
data_sales = pd.read_csv("100 Sales Records.csv")

# Ambil fitur numerik
numerical_features = ['Order_ID', 'Units_Sold', 'Unit_Price', 'Unit_Cost',
                      'Total_Revenue', 'Total_Cost', 'Total_Profit']
x = data_sales[numerical_features].values

# Normalisasi
x = StandardScaler().fit_transform(x)

# PCA ke 2 dimensi
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Tambahkan label kategori (misal: Item_Type)
principal_df['Item_Type'] = data_sales['Item_Type']

# Visualisasi PCA dengan warna
plt.figure(figsize=(10, 10))
targets = principal_df['Item_Type'].unique()
colors = plt.cm.tab20.colors  # Maksimal 20 warna unik

for target, color in zip(targets, colors):
    indices = principal_df['Item_Type'] == target
    plt.scatter(principal_df.loc[indices, 'PC1'],
                principal_df.loc[indices, 'PC2'],
                c=[color], label=target, s=50, alpha=0.7)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of 100 Sales Records Grouped by Item Type')
plt.legend(loc='best', fontsize=9)
plt.grid(True)
plt.show()
```
