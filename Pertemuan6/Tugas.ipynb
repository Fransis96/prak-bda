```python
import sqlalchemy as sql

conn = "mysql+pymysql://root:@localhost/tugas"
sql_engine = sql.create_engine(conn)

# Gunakan connect() untuk menjalankan query
with sql_engine.connect() as conn:
    for r in conn.execute(sql.text("SHOW DATABASES")):
        print(r[0])
```

```python
# Load
region.to_sql("region", con=sql_engine, if_exists='replace', index=False)
print("Load ke MySQL selesai")
```

```python
# Load data sales
sales.to_sql("sales", con=sql_engine, if_exists='replace', index=False)
print("Load ke MySQL selesai")
```

```python
query = """
SELECT region.Region, SUM(sales.Total_Profit) AS Total_Profit FROM sales
INNER JOIN region ON sales.Country = region.Country
GROUP BY region.Region;
"""

data = pd.read_sql(query, sql_engine)
```
