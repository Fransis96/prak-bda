```python
import os
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
# 1
brain_size = "https://raw.githubusercontent.com/Fransis96/Datasets/refs/heads/main/brain_size.csv"
data = pd.read_csv(brain_size, sep=';', na_values=".")
piq = data[['PIQ']].copy()  # Buat salinan eksplisit
```

```python
# 2
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    os.environ["LOKY_MAX_CPU_COUNT"] = "4"
    kmeans.fit(piq)
    inertia.append(kmeans.inertia_)

# Visualisasi Elbow Method
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# 3
optimal_k = 2  
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
piq['cluster'] = kmeans.fit_predict(piq)  # Menambahkan label kluster ke DataFrame
```

```python
# 4
plt.scatter(piq.index, piq['PIQ'], c=piq['cluster'], cmap='Set1', s=10)
plt.xlabel('Index')
plt.ylabel('PIQ')
plt.title(f'Clustering PIQ dengan k={optimal_k}')
plt.show()
```
