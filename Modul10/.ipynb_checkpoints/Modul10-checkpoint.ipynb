```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=3, random_state=0)
```

```python
y.shape
```

```python
X, y = make_blobs(n_samples=300, centers=1, cluster_std=0.90, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
import os

wcss = []
for i in range(1, 11):
    kmeans = KMeans (n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    os.environ["LOKY_MAX_CPU_COUNT"] = "4"
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range (1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans (n_clusters=2, init='k-means++', max_iter=300, n_init=18, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

**2. Hierarchical Clustering**
   - Kita panggil data yang ada di modul 9 yaitu brain_size.csv 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
brain_size = "https://raw.githubusercontent.com/Fransis96/Datasets/refs/heads/main/brain_size.csv"
data = pd.read_csv(brain_size, sep=';', na_values=".")
```

```python
data.head()
```

- Kita akan lakukan clustering untuk variabel ke 2 (FSIQ) dan variabel ke 3(VIQ) dan kita beri nama variabel X. Data kita rubah dulu dalam bentuk array 

```python
X = data.iloc[:, [2, 3]].values
```

```python
X
```

- Jumlah data sebanyak 40 dengan 2 variabel (kolom 2 dan 3). Dan selanjutkan kita buat dendrogram nya

```python
X.shape
```

```python
dendrogram = sch.dendrogram(sch.linkage (X, method='ward'))
```

- Selanjutnya kita dapat menggunakan notasi steno untuk menampilkan semua sampel milik Kategori sebagai warna tertentu.

> Akan ditampilkan 2 cluster dari data diatas.

```python
model = AgglomerativeClustering(n_clusters=2, linkage='ward')
model.fit(X)
labels = model.labels_
```

```python
plt.scatter(X [labels==0, 0], X [labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue')
#plt.scatter(X[Labels==2, 0], X[Labels==2, 1], s=50, marker='o', color='green')
#plt.scatter(X[Labels==3, 0], X [Labels==3, 1], s=50, marker='o', color='purple')
#plt.scatter(X[Labels==4, 0], X [Labels==4, 1], s=50, marker='o', color='orange')
plt.show()
```
