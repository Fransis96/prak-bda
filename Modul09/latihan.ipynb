#Persiapan Data

```python
import pandas as pd
import numpy as np
```

#Simulasi dataset

```python
np.random.seed(42)
data = {
'nama': [f'Orang_{i}' for i in range(1, 101)],
'usia': np.random.randint(20, 60, 100),
'penghasilan': np.random.normal(7000000, 1500000,100).astype(int),
'jenis_kelamin': np.random.choice(['Laki-laki', 'Perempuan'],100),
'wilayah': np.random.choice(['Utara', 'Selatan', 'Timur','Barat'], 100)
}
```

```python
df = pd.DataFrame(data)
df.head()
```

#Statistik Deskriptif
#Tampilkan statistik deskriptif (mean, median, standar deviasi, min, max) untukkolom penghasilan dan usia.

```python
df[['usia', 'penghasilan']].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#Visualisasi Distribusi Data
#Buat histogram untuk kolom penghasilan dan usia.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['penghasilan'], kde=True, bins=20,color='skyblue')
plt.title('Distribusi Penghasilan')
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['usia'], kde=True, bins=20, color='salmon')
plt.title('Distribusi Usia')
plt.tight_layout()
plt.show()
```

#Visualisasi Distribusi Data
#Boxplot Penghasilan berdasarkan Jenis Kelamin

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='jenis_kelamin', y='penghasilan', data=df)
plt.title('Boxplot Penghasilan Berdasarkan Jenis Kelamin')
plt.show()
```

#Visualisasi Hubungan Antar Variabel
#Scatter Plot Usia vs Penghasilan

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x='usia', y='penghasilan', hue='jenis_kelamin',
data=df)
plt.title('Scatter Plot Usia vs Penghasilan')
plt.show()
```

#Visualisasi Hubungan Antar Variabel
#Heatmap Korelasi Antar Variabel Numerik

```python
plt.figure(figsize=(5, 4))
sns.heatmap(df[['usia', 'penghasilan']].corr(), annot=True,
cmap='coolwarm')
plt.title('Korelasi Antar Variabel Numerik')
plt.show()
```

#Penyajian Data Kategori
#Pie Chart Jenis Kelamin

```python
plt.figure(figsize=(5, 5))
df['jenis_kelamin'].value_counts().plot.pie(autopct='%1.1f%%',
startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Proporsi Jenis Kelamin')
plt.ylabel('')
plt.show()
```

#Penyajian Data Kategori
#Bar Chart Penghasilan Rata-rata per Wilayah

```python
mean_income_region = df.groupby('wilayah')['penghasilan'].mean().sort_values()
plt.figure(figsize=(7, 5))
sns.barplot(x=mean_income_region.index,y=mean_income_region.values, palette='viridis')
plt.title('Penghasilan Rata-rata per Wilayah')
plt.ylabel('Penghasilan')
plt.show()
```
