### Setelah itu kita buat coding untuk rumus deviasi standart

```python
def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums)
    for n in nums:
        diffs += (n - avg)**(2)
    return (diffs/(len(nums)-1))**(0.5)
```

Misal akan dicari nilai deviasi standart dari variabel VIQ dengan cara membuat inisiasi variabel tersebut dan setelah itu kita berikan fungsi deviasi standar yang sdh didefinisikan.

```python
viq = df['VIQ']
viq
```

```python
print(stdev(viq))
```

```python
data_height = df['Height']
data_height
print(stdev(data_height))
```

```python
data_height.isna().sum()
```

```python
data_height = data_height.fillna(data_height.mean())

print(stdev(data_height))
```
